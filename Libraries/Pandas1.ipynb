## Multi-Index and Index Hierarchy

```python
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside, inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

## Dealing with missing data

```python
df = pd.DataFrame({'A' : [1,2,np.nan],
                   'B' : [5,np.nan,np.nan],
                   'C' : [1,2,3]})
```

```python
df
```

```python
# It drops every row that has NaN values
df.dropna()
```

```python
# it drops every column that hs NaN values
df.dropna(axis  = 1)
```

```python
# it sets a threshhold to the amount of NaN values
df.dropna(thresh = 2)
```

```python
#To fill the Nan values with a another value
df.fillna(value = 'Fill value')
```

```python
# Here we take the mean of the column and replace all Nan values with that value
df['A'].fillna(value = df['A'].mean())
```

## Group by

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
by_comp =  df.groupby("Company")
```

```python
by_comp['Sales'].mean()
```

```python
by_comp['Sales'].std()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```
